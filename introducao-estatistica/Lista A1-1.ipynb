```r
# 1.1 COnstrua um vetor no R com os valores gastos
valoresGastos <- c(450, 300, 380, 320, 500, 480, 420, 400, 350, 370, 410)
valoresGastos
```

```r
# 1.2 Determine os valores máximo e mínimo
maximoValor <- max(valoresGastos)
maximoValor

minimoValor <- min(valoresGastos)
minimoValor
```

```r
# 1.3 Determine em quais meses ocorreram os valores máximos e mínimos
which(valoresGastos == maximoValor)
#positivos <- ex1 >= 0

which(valoresGastos == minimoValor)
```

```r
# 1.4 Determine o número de meses em que se gastou acima da média
mediaGastos = mean(valoresGastos)
mediaGastos

acimaMedia <- valoresGastos > mediaGastos
acimaMedia

length(which(acimaMedia == TRUE))
```

```r
# 1.5 Determine o percentual que cada um desses meses representou em relacao ao total anual
somaGastos <- sum(valoresGastos)
somaGastos

percentual <- (valoresGastos / somaGastos) * 100
percentual
```

```r
# Exercício 2
set.seed(1)
notas <- round(runif(100) * 100)
notas
```

```r
# 2.1 Explique cada elemento deste comando.
# runif(100) - Gera uma amostra de tamanho 100 randômica, que vai de 0 a 1 (que são valores default da função)
runif(100)
help(runif)

# round - arredonda o valor passado como argumento para um número específico de casas decimais (default é 0).
round(1.5)
help(round)
```

```r
# 2.2 Crie um vetor chamado conceito, que contém os conceitos correspondentes às notas de cada aluno
conceito <- rep('', 100)

conceito[notas >= 90] = 'A'
conceito[notas >= 80 & notas <= 89] = 'B'
conceito[notas >= 70 & notas <= 79] = 'C'
conceito[notas >= 60 & notas <= 69] = 'D'
conceito[notas >= 40 & notas <= 59] = 'E'
conceito[notas <= 40] = 'F'
conceito
```

```r
# 2.3 Crie um vetor chamado resultado, que contém a situação final do aluno (APROVADO ou REPROVADO)
resultado <- rep('', 100)

resultado[notas >= 60] = 'APROVADO'
resultado[notas < 60] = 'REPROVADO'
resultado
```
